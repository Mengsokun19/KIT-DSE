```python
import matplotlib.pyplot as plt
x = [1, 2, 3, 4, 5, 6, 7, 8]
y1 = [1, 4, 9, 16, 25, 36, 49,64]
y2 = [1, 4, 6, 8, 10, 12, 14, 16]
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
l1 = ax.plot(x,y1,'ys-') # solid line with yellow colour and square marker
l2 = ax.plot(x,y2,'go--') # dash line with green colour and circle marker
ax.legend(labels = ('x^2', '2x'), loc = 'lower right') # legend placed at lower right
ax.set_title("Demo Line Plot")
ax.set_xlabel('x values')
ax.set_ylabel('y values')
plt.show()
```

```python
import matplotlib.pyplot as plt
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax = fig.add_subplot(2,1,1)
ax.plot(x,y1,'ys-')
ax = fig.add_subplot(2,1,2, facecolor='c')
ax.plot(x,y2,'ro--')
```

```python
import matplotlib.pyplot as plt
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax = fig.add_subplot(1,2,1)
ax.plot(x,y1,'ys-')
ax = fig.add_subplot(1,2,2, facecolor='c')
ax.plot(x,y2,'ro--')
```

```python
fig, axs = plt.subplots(2, 2)
plt.tight_layout()
axs[0, 0].plot(x, y1)
axs[0, 0].set_title('Axis [0, 0]')
axs[0, 1].plot(x, y1, 'tab:orange')
axs[0, 1].set_title('Axis [0, 1]')
axs[1, 0].plot(x, y2, 'tab:green')
axs[1, 0].set_title('Axis [1, 0]')
axs[1, 1].plot(y1, y2, 'tab:red')
axs[1, 1].set_title('Axis [1, 1]')
```

```python
import pandas as pd
# Import the csv file
df = pd.read_csv("https://raw.githubusercontent.com/stulasiprasad/KIT-DSE/main/anscombe.csv")
print(df)
```

```python
import numpy as np
for x in (df['x1'], df['x2'], df['x3'], df['x4']):
    print(np.mean(x), np.round(np.std(x),2))
```

We see that all the data sets have the same mean and standard deviation for `x`.

```python
for y in (df['y1'], df['y2'], df['y3'], df['y4']):
    print(np.round(np.mean(y),1), np.round(np.std(y),1))
```

Similarly for `y`: all the data sets have the same mean and standard deviation.

```python
for x, y in ((df['x1'], df['y1']), (df['x2'], df['y2']), (df['x3'], df['y3']),  (df['x4'], df['y4'])):
    a, b = np.polyfit(x, y, 1)
    print(np.round(a,1), np.round(b,1))
```

All the coefficients are the same
